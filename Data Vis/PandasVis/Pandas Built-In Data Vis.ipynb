# Pandas plots
Pandas plots can be called off of the dataframe object or column of dataframe

This is useful for quick, basic plots.

But if you want to customize, should use seaborn or matplotlib.

2 methods:

- .plot(kind='hist') 

- .plot.hist(...)

Note: In general, arguments from matplotlib also apply to Pandas plot methods (e.g., figsize, linewidth (lw), colormap (cmap), etc)

# Plot Types

There are several plot types built-in to pandas, most of them statistical plots by nature:

* df.plot.area     
* df.plot.barh     
* df.plot.density  
* df.plot.hist     
* df.plot.line     
* df.plot.scatter
* df.plot.bar      
* df.plot.box      
* df.plot.hexbin   
* df.plot.kde      
* df.plot.pie

You can also just call df.plot(kind='hist') or replace that kind argument with any of the key terms shown in the list above (e.g. 'box','barh', etc..)
___

### Histogram


```python
df1['A'].plot(kind='hist')
```

```python
df1['A'].plot.hist(bins=30)   # this is Pandas built-in hist method. 
# it's built on matplotlib, so can specify matplotlib or seaborn arguments (make sure those libraries are imported)
# Note: style is "nicer" because seaborn has been imported (see 2nd cell)
```

### Area

```python
df2.plot.area()
```

```python
df2.plot.area(alpha=0.4)
```

### Bar

```python
df2.plot.bar()
```

```python
df2.plot.bar(stacked=True)
```

### Line

```python
df1.plot.line(x=df1.index,y='B', figsize=(12,3), lw=1)
```

### Scatter

```python
df1.plot.scatter(x='A', y='B', c='C', cmap='coolwarm') # c is color variable (optional)  

# with a colormap
```

```python
df1.plot.scatter(x='A', y='B', s=df1['C']*100) # s is sizing
# with sizing
```

### Boxplot

```python
df2.plot.box()
```

### Hexagonal Bin 
For bivariate data

```python
my_df = pd.DataFrame(np.random.randn(1000,2),columns=['a','b'])
my_df.head()
```

```python
my_df.plot.hexbin(x='a',y='b',gridsize=25, cmap='coolwarm')
```

### KDE

```python
df2['a'].plot.density() #same as  .plot.kde()
```
