```python
# check each permutation (fit into an equation of all 2-digit numbers, like ab - cd = ef)

# get all permutations
all_permuts = list(itertools.permutations([6,5,4,3,2,1]))

# for each permutation, check that ab - cd - ef = 0 (which is equivalent to checking ab - cd = ef)
for x in all_permuts:
    check_x = (int(str(x[0] )+str(x[1] )) - int(str(x[2] )+str(x[3] )) -int(str(x[4] )+str(x[5] )))
    print (x, " " + str(x[0] )+str(x[1] )+" - " + str(x[2] )+str(x[3] )+" = "+str(x[4] )+str(x[5] )+" ",  check_x==0)
        
```
